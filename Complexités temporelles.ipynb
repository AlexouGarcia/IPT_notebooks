```python
import matplotlib.pyplot as plt
import time
```

```python
def moyenne(L):
    tdebut = time.time()
    somme = 0
    for i in L:
        somme +=i
    somme = somme/len(L)
    tfin = time.time()
    temps_calcul = tfin-tdebut
    return somme,temps_calcul

def variance_1(L):
    tdebut = time.time()
    somme = 0
    
    for i in L:
        somme +=(i-moyenne(L)[0])**2 # moyenne(L)[0] est la moyenne et moyenne(L)[1] le temps de calcul associé
    somme = somme/len(L)
    tfin = time.time()
    temps_calcul = tfin-tdebut
    return somme,temps_calcul

def variance_2(L):
    tdebut = time.time()
    somme = 0
    mu = moyenne(L)[0]
    for i in L:
        somme +=(i-mu)**2
    somme = somme/len(L)
    tfin = time.time()
    temps_calcul = tfin-tdebut
    return somme,temps_calcul
```

```python
taille_L = [i for i in range(10,10**3)]
temps_moyenne = []
temps_variance_1 = []
temps_variance_2 = []

for n in taille_L:
    L = [i for i in range(n)] # L est une liste de taille n
    valeur_moyenne,temps_calcul_moyenne = moyenne(L)
    temps_moyenne += [temps_calcul_moyenne]
    valeur_variance_1,temps_calcul_variance_1 = variance_1(L)
    temps_variance_1 += [temps_calcul_variance_1]
    valeur_variance_2,temps_calcul_variance_2 = variance_2(L)
    temps_variance_2 += [temps_calcul_variance_2]
```

```python
plt.plot(taille_L,temps_moyenne)
plt.xlabel('Taille de liste L')
plt.ylabel('Temps de calcul (s)')
plt.title('Temps de calcul de la moyenne en fonction de la taille de la liste L')
```

```python
plt.plot(taille_L,temps_variance_1)
plt.xlabel('Taille de liste L')
plt.ylabel('Temps de calcul (s)')
plt.title('Temps de calcul de la variance (méthode 1) en fonction de la taille de la liste L')
```

```python
plt.plot(taille_L,temps_variance_2)
plt.xlabel('Taille de liste L')
plt.ylabel('Temps de calcul (s)')
plt.title('Temps de calcul de la variance (méthode 2) en fonction de la taille de la liste L')
```

```python
# superposition des courbes :
plt.plot(taille_L,temps_variance_1)
plt.plot(taille_L,temps_variance_2)
plt.xlabel('Taille de liste L')
plt.ylabel('Temps de calcul (s)')
plt.title('Temps de calcul de la variance en fonction de la taille de la liste L')
plt.legend(['Méthode 1','Méthode 2'])
```
