```python
import matplotlib.pyplot as plt
import numpy as np

n = 1000
xtrue = np.linspace(0,np.pi/2,n+1)
ytrue = np.cos(xtrue)

n = 4
x = np.linspace(0,np.pi/2,n+1)
y = np.cos(x)
```

```python
plt.plot(xtrue,ytrue)
```

```python
# Méthode des rectangles à gauche:
def f(x):
    return np.cos(x)

X_rect_gauche = []
Y_rect_gauche = []
for i in range(n):
    X_rect_gauche += [x[i],x[i],x[i+1]]
    Y_rect_gauche += [0,f(x[i]),f(x[i])]
Y_rect_gauche += [0]
X_rect_gauche += [x[i+1]]
```

```python
plt.plot(X_rect_gauche,Y_rect_gauche)
plt.hold(1)
plt.plot(xtrue,ytrue)
```

```python
# Méthode des rectangles à droite:

X_rect_droite = []
Y_rect_droite = []
for i in range(n):
    X_rect_droite += [x[i],x[i],x[i+1]]
    Y_rect_droite += [0,f(x[i+1]),f(x[i+1])]
Y_rect_droite += [0]
X_rect_droite += [x[i+1]]
```

```python
plt.plot(X_rect_droite,Y_rect_droite)
plt.hold(1)
plt.plot(xtrue,ytrue)
```

```python
# Méthode des rectangles milieu:

X_rect_milieu = []
Y_rect_milieu = []
for i in range(n):
    X_rect_milieu += [x[i],x[i],x[i+1]]
    Y_rect_milieu += [0,f((x[i]+x[i+1])/2),f((x[i]+x[i+1])/2)]
Y_rect_milieu += [0]
X_rect_milieu += [x[i+1]]
```

```python
plt.plot(X_rect_milieu,Y_rect_milieu)
plt.hold(1)
plt.plot(xtrue,ytrue)
```

```python
# Méthode des trapèzes:

X_trap = []
Y_trap = []
for i in range(n):
    X_trap += [x[i],x[i],x[i+1]]
    Y_trap += [0,f(x[i]),f(x[i+1])]
Y_trap += [0]
X_trap += [x[i+1]]
```

```python
plt.plot(X_trap,Y_trap)
plt.hold(1)
plt.plot(xtrue,ytrue)
```
